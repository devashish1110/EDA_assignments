##Importing the Dataframe

##Understanding the difference between append,concat, merge and join.

```python
df1 = df.head(10)
df2 = df.head(15)
appended_df = pd.concat([df1, df2], ignore_index=True)
print(appended_df)
#This is append because it is row-wise stacking. The values are appended at the end of the first table's values.
```

```python
appended_df = pd.concat([df1, df2], ignore_index=False)
print(appended_df)
#This is also append only but the indexes dont continue. The second tabel is added as is.
#ignore_index = False is by default
```

```python
appended_df = pd.concat([df1,df2], axis = 0)
print(appended_df)
#axis = 0 is also by default, same as ignore_index = False.
```

```python
concat_df = pd.concat([df1,df2], axis = 1,ignore_index=False)
print(concat_df)
#Now this is concatenation. Columnwise stacking
```

###Combining the Dataset

```python
#Before combining creating the datasets
df_work = pd.DataFrame({
    "rownames": df["rownames"].head(8),
    "Work_Hours": df["work"].head(8)
})
df_age = pd.DataFrame({
    "rownames": df["rownames"].head(8),
    "Age": df["age"].head(8)
})
```

```python
df_work
```

```python
df_age
```

```python
df_combined = pd.concat([df_work, df_age], axis=1)
print(df_combined)
```

###Inner Join

```python
df = df_work.merge(df_age, how = 'inner')
df
#values of both work and age will appear here.
```

###Left Join

```python
df = df_work.merge(df_age, how = 'left')
df
#preserves the values in the left df.
```

###Right Join

```python
df = df_work.merge(df_age, how = 'right')
df
```

###Outer Join

```python
df = df_work.merge(df_age, how = 'outer')
df
#union of both the df's.
```

```python
stacked = df_summary.stack()
stacked
#stack() rotates from col to row
```

##Discretizatoin and Binning

```python
df['work'] = df['work'].replace(np.nan, 0)
```

```python
bins = [0, 20, 40, 60]
bin_labels = ['low','medium', 'high']
```

```python
df['work_category'] = pd.cut(df['work'],bins = bins, labels=bin_labels, include_lowest=True)
df
print(df[['work', 'work_category']])
```

```python
print(pd.value_counts(df['work_category']))
```

```python
df['work_category_auto'] = pd.cut(df['work'], 5, precision=1)
print(df[['work', 'work_category_auto']])
```

```python
df_nonan = df['work'].dropna()
df.loc[df['work'].notna(), 'work_category_quantile'] = pd.qcut(df_nonan, 4, duplicates='drop')
print(df[['work', 'work_category_quantile']])
```
