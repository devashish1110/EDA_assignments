```python
import pandas as pd
import numpy as np
```

##Creating the Dataframes

```python
df1 = pd.DataFrame({
    'EmployeeID': [201, 203, 205, 207, 209, 211, 213, 215, 217, 219, 221, 223, 225, 227, 229],
    'MonthlySales': [5200, 4300, 6100, 7500, 4900, 5600, 5800, 6000, 7200, 8100, 6900, 5300, 6700, 7600, 8200]
})

df2 = pd.DataFrame({
    'EmployeeID': [202, 204, 206, 208, 210, 212, 214, 216, 218, 220, 222, 224, 226, 228, 230],
    'MonthlySales': [6400, 5800, 5100, 7300, 6900, 7100, 6600, 5900, 7400, 8100, 7000, 6200, 6800, 7500, 7700]
})

print(df1)
print(df2)
```

##Understanding the difference between append,concat, merge and join.

```python
df = pd.concat([df1, df2], ignore_index=True)
print(df)
#This is append because it is row-wise stacking. The values are appended at the end of the first table's values.
```

```python
df = pd.concat([df1, df2], ignore_index=False)
print(df)
#This is also append only but the indexes dont continue. The second tabel is added as is.
#ignore_index = False is by default
```

```python
df = pd.concat([df1,df2], axis = 0)
print(df)
#axis = 0 is also by default, same as ignore_index = False.
```

```python
df = pd.concat([df1,df2], axis = 1,ignore_index=False)
print(df)
#Now this is concatenation. Columnwise stacking
```

#Merging





```python
df1Elec = pd.DataFrame({
    'ProductID': [101, 103, 105, 107, 109, 111, 113, 115, 117, 119, 121],
    'Sales_Electronics': [220, 340, 410, 295, 510, 630, 410, 375, 490, 560, 600]
})
df2Elec = pd.DataFrame({
    'ProductID': [102, 104, 106, 108, 110, 112, 114, 116, 118, 120, 122, 124, 126, 128, 130],
    'Sales_Electronics': [430, 520, 390, 610, 470, 530, 440, 580, 495, 615, 555, 505, 460, 540, 590]
})
df1Furn = pd.DataFrame({
    'ProductID': [101, 103, 105, 107, 109, 111, 113, 115, 117, 119, 121, 123, 125, 127, 129],
    'Sales_Furniture': [150, 210, 190, 260, 300, 280, 230, 310, 275, 340, 295, 260, 330, 240, 320]
})
df2Furn = pd.DataFrame({
    'ProductID': [102, 104, 106, 108, 110, 112, 114, 116, 118, 120],
    'Sales_Furniture': [200, 250, 230, 290, 310, 270, 260, 300, 280, 350]
})
```

```python
df1Elec
```

```python
df2Elec
```

```python
df1Furn
```

```python
df2Furn
```

###Combining the Dataset

```python
dfElec = pd.concat([df1Elec,df2Elec], ignore_index = True)
dfFurn = pd.concat([df1Furn,df2Furn], ignore_index = True)
df = pd.concat([dfElec,dfFurn], axis = 1)
df
```

###Inner Join

```python
df = dfElec.merge(dfFurn, how = 'inner')
df
#Sales of both Electronics and Furnitures will appear in this.
```

###Left Join

```python
df = dfElec.merge(dfFurn, how = 'left')
df
#preserves the values in the left df.
```

###Right Join

```python
df = dfElec.merge(dfFurn, how = 'right')
df
```

###Outer Join

```python
df = dfElec.merge(dfFurn, how = 'outer')
df
#union of both the df's.
```

##Discretizatoin and Binning

```python
weights = [35, 37, 40, 42, 38, 39, 50, 46, 55, 49, 47, 43]
weights
```

```python
bins = [34,40,45,50,55]
category = pd.cut(weights,bins)
category
```

```python
category = pd.cut(weights,bins,right = False)
category
```

```python
pd.value_counts(category)
```

```python
bin_names = ['low', 'average', 'high', 'very high']
category = pd.cut(weights,bins,labels = bin_names)
category
```

```python
category = pd.cut(weights, 5 , precision = 1)
category
#makes its own equi length bins. precision is for upto a certain amount of decimal places.
```

```python
random_no = np.random.rand(20)
random_no
```

#####cut into quartiles
